# ESTIMATED MARKET SIZE ANALYSIS OF ELECTRIC VEHICLES IN THE UNITED STATES

```python
#NOW ,LET'S MOVE FORWARD TORWARDS FINFINGTHE ESTIMATED  MARKET SIZE  OF ELECTRIC VEHICLES IN THE UNITED STATES.WE'LL FIRST COUNT THE NUMBER OF OF EV REGESTEREDEVRY YEAR
```

```python
ev_registration_counts = ev_data["Model Year"].value_counts().sort_index()
ev_registration_counts
```

```python
#THE ABOVE DATA SHOWS THE PROGRESSIVE UPTAKE OF EVS OVER THE YEARS BUT TAKE NOTE OF 2024 WHWER THE DATA WAS CAPTURED TILL MARCH
```

```python
#To forecast the total number of EVs expected to be registered in 2024, we can use a growth rate based approachfrom previous complete years.
```

```python
#Weâ€™ll calculate the Compound Annual Growth Rate (CAGR) between a recent year with complete data (2023) andan earlier year to project the 2024 figures. Additionally, using this growth rate, we can estimate the market size forthe next five years. Letâ€™s proceed with these calculations:
```

```python
from scipy.optimize import curve_fit
import numpy as np
```

```python
#FILTER THE DATASET TO  INCLUDE YEARS  WITH COMPLETE DATA,ASSUMING  2023 IS THE LAST COMPLETE YEAR
```

```python
filtered_years = ev_registration_counts[ev_registration_counts.index<=2023]
filtered_years
```

```python
filtered_years.index
```

```python
filtered_years.index.min()
```

```python
#DEFINE A FUNCTION FOR EXPONENTIAL GROWTH  TO FIT THE DATA
```

```python
def exp_growth(x,a,b):
    return a *np.exp(b*x)
```

```python
#PREPARE DATA FOR  CURVE FITTING
```

```python
x_data = filtered_years.index- filtered_years.index.min()
y_data = filtered_years.values
```

```python
x_data
```

```python
#FIT THE DATA  TO THE EXPONENTIAL GROWTH FUNCTION
```

```python
params,covarience = curve_fit(exp_growth,x_data,y_data)
```

```python
#USE THE FITTED FUNCTION TO FORECAST THE NUMBER OF EVS FOR THE 2024 AND THE NEXT FIVE YEARS
```

```python
forecast_years = np.arange(2024,2024 + 6) - filtered_years.index.min()
forecasted_values = exp_growth(forecast_years,*params)
```

```python
#CREATE A DICTIONARY TO DISPLAY THE FORECASTED VALUES FOR  EASIER INTERPRETATION
```

```python
forecasted_evs = dict(zip(forecast_years + filtered_years.index.min(),forecasted_values))
print(forecasted_evs)
                      
```

```python
#NOW LETS PLOT THE ESTIMATED MARKET SIZE DATA
```

```python
#PREPARE DATA FOR PLOTTING
```

```python
years = np.arange(filtered_years.index.min(),2029 + 1)
actual_years = filtered_years.index
forecast_years_full = np.arange(2024,2029 + 1)
```

```python
#ACTUAL AND FORECASTED VALUES
```

```python
actual_values = filtered_years.values
forecasted_values_full =  [forecasted_evs[year] for year in forecast_years_full]
plt.figure(figsize=(12,8))
plt.plot(actual_years,actual_values,"bo-",label = "Actual Registrations")
plt.plot(forecast_years_full,forecasted_values_full,"ro--",label = "forecasted Registration")
plt.title("Current & Estimated EV Market")
plt.xlabel("Year")
plt.ylabel("Number of EV Registration")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#From the above graph, we can see:
#The number of actual EV registrations remained relatively low and stable until around 2010, after which therewas a consistent and steep upward trend, suggesting a significant increase in EV adoption.
#The forecasted EV registrations predict an even more dramatic increase in the near future, with the numberof registrations expected to rise sharply in the coming years.
```
